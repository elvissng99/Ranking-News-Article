```python
text_file = open("swedish-stopwords.txt", "r", encoding='utf-8')
stopwords_git = text_file.read().splitlines()+ ["år","se","just","sävsjö"]
stopwords_nltk = stopwords.words('swedish')

stopwords = list(dict.fromkeys(stopwords_git + stopwords_nltk))
```

```python
def un_html(text):
    text = re.sub(re.compile('<.*?>'), '', text)
    text = html.unescape(text)
    return text
```

```python
def remove_numbers(doc):
    doc = re.sub(r'[0-9]', '', doc)
    return doc
```

```python
import spacy

nlp = spacy.load("sv_core_news_sm")

def process(df):
    #HTML
    df = df.apply(lambda text: [un_html(x) for x in text]) 
    #Numbers
    df = df.apply(lambda text: [remove_numbers(x) for x in text]) 
    #lowecase
    df['Text'] = df['Text'].apply(lambda text: text.lower()) 
    #remove punctions, , . ! ?
    df['Text'] = df["Text"].apply(lambda text: re.sub(r'[^\w\s]','',text))                                       
    #lowecase
    df['Text'] = df['Text'].str.split()
    #stopwords
    df['Text'] = df['Text'].apply(lambda text: " ".join([word for word in text if word not in stopwords]))
    
    #df['join_wo_stopwords'] = df['removed_stop_words'].apply(lambda text: " ".join(text))

    df['Text'] = df['Text'].apply(lambda text: " ".join([word.lemma_ for word in nlp(text)]))

    #df['lemma'] = df['lemma'].apply(lambda text: " ".join(text))

    return pd.DataFrame(df['Text'])
```

```python
def create_df(df_articles):
    #get text that need pre processing
    df = df_articles.drop(columns="Title")

    df_pro = process(df)
    #join back with titles
    df_true = df_pro.join(df_articles["Title"]).reset_index()
    #remove text that are short
    df_true = df_true[df_true["Text"].apply(lambda text: len(text.split()) > 50)]

    return df_true.reset_index(drop = True)
```

```python
dataframe = create_df(df_art)
```

```python
# Create Document - Topic Matrix
def get_df_document_topics(best_lda_model):
    lda_output = best_lda_model.transform(data_vectorized)

    # column names
    topicnames = ["Topic" + str(i) for i in range(best_lda_model.n_components)]

    # index names
    docnames = ["Doc" + str(i) for i in range(len(dataframe))]

    # Make the pandas dataframe
    df_document_topic = pd.DataFrame(np.round(lda_output, 2), columns=topicnames, index=docnames)

    # Get dominant topic for each document
    dominant_topic = np.argmax(df_document_topic.values, axis=1)
    df_document_topic['dominant_topic'] = dominant_topic

    # Styling
    def color_green(val):
        color = 'green' if val > .1 else 'black'
        return 'color: {col}'.format(col=color)

    def make_bold(val):
        weight = 700 if val > .1 else 400
        return 'font-weight: {weight}'.format(weight=weight)

    # Apply Style
    df_document_topics = df_document_topic.head(15).style.applymap(color_green).applymap(make_bold)
    return df_document_topic,topicnames, lda_output

#df_document_topic, topic_names,lda_output = get_df_document_topics(best_lda_model)
```

```python
def get_topic_ditrubution(document_topic):

    df_topic_distribution = document_topic['dominant_topic'].value_counts().reset_index(name="Num Documents")
    df_topic_distribution.columns = ['Topic Num', 'Num Documents']

    return df_topic_distribution

#df_topic_distribution = get_topic_ditrubution(df_document_topic)
#df_topic_distribution
```

```python
# Topic-Keyword Matrix
def get_keywords(topicnames, model):
    df_topic_keywords = pd.DataFrame(model.components_)

    # Assign Column and Index
    df_topic_keywords.columns = vectorizer.get_feature_names()
    df_topic_keywords.index = topicnames
    return df_topic_keywords

# View
```

```python
# Show top n keywords for each topic
def show_topics(model,vectorizer=vectorizer,  n_words=20):
    keywords = np.array(vectorizer.get_feature_names())
    topic_keywords = []
    for topic_weights in model.components_:
        top_keyword_locs = (-topic_weights).argsort()[:n_words]
        topic_keywords.append(keywords.take(top_keyword_locs))
    return topic_keywords

#topic_keywords = show_topics(vectorizer=vectorizer, lda_model=best_lda_model, n_words=15)        

def get_topic_keywords(topic_keywords):
    # Topic - Keywords Dataframe
    df_topic_keywords = pd.DataFrame(topic_keywords)
    df_topic_keywords.columns = ['Word '+str(i) for i in range(df_topic_keywords.shape[1])]
    df_topic_keywords.index = ['Topic '+str(i) for i in range(df_topic_keywords.shape[0])]

    return df_topic_keywords
    
#df_topic_keywords = get_topic_keywords(topic_keywords)
```

```python
# Define function to predict topic for a given text document.
#nlp = spacy.load('en', disable=['parser', 'ner'])

def predict_topic(model,text,df_topic_keywords, nlp=nlp):

    mytext_4 = vectorizer.transform(text)

    # Step 4: LDA Transform
    topic_probability_scores = model.transform(mytext_4)
    topic = df_topic_keywords.iloc[np.argmax(topic_probability_scores), :].values.tolist()
    return topic, topic_probability_scores

# Predict the topic
#mytext = [dataframe['Text'][10]]
#topic, prob_scores = predict_topic(text = mytext)
#print(topic,prob_scores)
```

```python
def add_topics_to_df(model,df,topic_keywords):
    list_of = []
    df_topic_keywords = get_topic_keywords(topic_keywords)
    
    for idx,text in enumerate(dataframe['Text']):
        topic, prob_scores = predict_topic(model,text = [text],df_topic_keywords=df_topic_keywords)
        list_of.append(np.argmax(prob_scores))

    df_topic = dataframe.join(pd.DataFrame(list_of, columns=["idx"]))

    return df_topic
```

```python
# Construct the k-means clusters
from sklearn.cluster import KMeans
from sklearn.decomposition import LatentDirichletAllocation, TruncatedSVD

def cluster_plot(lda_output, n_clusters):

    clusters = KMeans(n_clusters=n_clusters, random_state=100).fit_predict(lda_output)

    # Build the Singular Value Decomposition(SVD) model
    svd_model = TruncatedSVD(n_components=2)  # 2 components
    lda_output_svd = svd_model.fit_transform(lda_output)

    # X and Y axes of the plot using SVD decomposition
    x = lda_output_svd[:, 0]
    y = lda_output_svd[:, 1]

    # Weights for the 15 columns of lda_output, for each component
    print("Component's weights: \n", np.round(svd_model.components_, 2))

    # Percentage of total information in 'lda_output' explained by the two components
    print("Perc of Variance Explained: \n", np.round(svd_model.explained_variance_ratio_, 2))

    plt.figure(figsize=(12, 12))
    plt.scatter(x, y, c=clusters)
    plt.xlabel('Component 2')
    plt.xlabel('Component 1')
    plt.title("Segregation of Topic Clusters", )
```
